### **Construya una tabla que contenga _c1 y una lista separada por ':' de los valores de la columna _c2 para el archivo `tbl0.tsv`.**

    Rta/
                                   _c1
      _c0
    0   A              1:1:2:3:6:7:8:9
    1   B                1:3:4:5:6:8:9
    2   C                    0:5:6:7:9
    3   D                  1:2:3:5:5:7
    4   E  1:1:2:3:3:4:5:5:5:6:7:8:8:9

```python

Resultado = tbl0.copy()
Resultado = Resultado.sort_values(['_c1', '_c2', '_c0'], ascending=True)
Resultado_10 = Resultado.groupby('_c1')['_c2'].agg(lambda x: ':'.join(x.astype(str))).reset_index()
Resultado_10.set_index('_c1', inplace=True)
Resultado_10
```

### **Construya una tabla que contenga _c0 y una lista separada por ',' de los valores de la columna _c4 del archivo `tbl1.tsv`.**

### **Construya una tabla que contenga _c0 y una lista separada por ',' de los valores de la columna _c5a y _c5b (unidos por ':') de la tabla `tbl2.tsv`.**

    Rta/
        _c0                                  _c5
    0     0        bbb:0,ddd:9,ggg:8,hhh:2,jjj:3
    1     1              aaa:3,ccc:2,ddd:0,hhh:9
    2     2              ccc:6,ddd:2,ggg:5,jjj:1
    ...
    37   37                    eee:0,fff:2,hhh:6
    38   38                    eee:0,fff:9,iii:2
    39   39                    ggg:3,hhh:8,jjj:5

```python
Respuesta_11 = tbl2.copy()
Respuesta_11_T = Respuesta_11.sort_values(['_c5a','_c5b', '_c0'], ascending=True)
Respuesta_To = Respuesta_11_T.assign( _c5 = Respuesta_11_T['_c5a']+":"+Respuesta_11_T['_c5b'].astype(str))
Respuesta_Total = Respuesta_To.groupby('_c0')['_c5'].agg(','.join).reset_index()

Respuesta_Total
```
